```python
df.drop(["Univ"],axis=1,inplace=True)
```

```python
a=df.values
```

```python
a
```

```python
stscaler=StandardScaler().fit(a)
x=stscaler.transform(a)
```

```python
dbscan=DBSCAN(eps=1,min_samples=5)
dbscan.fit(x)
```

```python
x
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([df,cl],axis=1)
```

```python
from operator import sub
fig=plt.figure(figsize=(5,7))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(df['SAT'],df['Top10'],df['SFRatio'],c=dbscan.labels_)
ax.set_xlabel("SAT")
ax.set_ylabel("Top10")
ax.set_zlabel("SFRatio")
ax.set_title("3D Graph")
plt.show()
```
